```python
#Individua lo spettro che meglio approssima lo spettro mediano della popolazione
#Permette di eseguire la verifica sullo spettro totale o su porzioni di esso
#accetta file di output da Strata formato .csv
#versione 13/01/2018
import os
from pandas import DataFrame
import numpy as np
import math
import pylab as pl
```

```python
a = 'C:\DOTTORATO\MICROZONAZIONE_SISMICA\_Python\INTEGRALE_FH\FromStrataCSV' #input(Inserisci Percorso File: ) - Persorso file
b = '.csv'		# Input('Inserisci Estensione File: ') - Estensione file
doelabo = [] 	# Creo un array per i file su i quali devo eseguire i calcoli
nodoelabo = [] 	# Creo un array per i file su i quali NON devo eseguire i calcoli
```

```python
def leggifile(direct,estens):
    for f in os.listdir(direct): 	# Iterazione su tutti i file della cartella
        if f.endswith(estens): 		# Verifico l'estensione corretta
            doelabo.append(f)
        else:
            nodoelabo.append(f)
```

```python
leggifile(a,b)
```

```python
print 'il file che contiene gli spettri di output è: %s'% doelabo[0]
```

```python
dati = open(doelabo[0],'r' ).readlines()				# Accedo al file csv
titolinfo = [ i.strip().split(',') for i in dati[:3] ]	# Memorizzo le info
zout = [ i.strip().split(',') for i in dati[3:] ]		# Legge tutti i record
zout = np.array( zout, dtype=np.float )					# Dichiaro la tipologia dei dati
zout = DataFrame(zout)									# Costruisco il dataframe
naccel2 = len(zout.columns) - 2 						# Difinisce il numero di Accelerogrammi per output stampe
Lquadro = [] 											# Array valori scarto
naccel = len(zout)
```

```python
limiteinf = 0.0						#Limite inferiore verifica
limitesup = 0.8						#Limite superiore verifica
	
for vala in range(1,naccel2,1): 	#naccel2
	Lquadroi = []
	for val in range(0,naccel,1):	#visualizzo gli spettri
		if float(zout[0][val]) > limiteinf and float(zout[0][val]) < limitesup:
			x = math.log(zout[vala][val])
			y = math.log(zout[naccel2][val])
			z = math.pow((x-y),2)
			Lquadroi.append(z)
	Lquadro.append(sum(Lquadroi)) 	#Popolo l'array Lquadroi 

bestacc = str(Lquadro.index(min(Lquadro)))
#print bestacc
print len(Lquadro)
print '*** Selezionare accellerogramma SITE: %s ***'% bestacc
#non confondere con il numero sequenziale di Strata!!!*

pl.plot(zout[0],zout[naccel2],'r-',lw=1.5,label='Median')
pl.plot(zout[0],zout[int(bestacc)+1],'b-',lw=1.5,label='Best-Fit')
pl.minorticks_on()
pl.title(titolinfo[2][int(bestacc)])
pl.grid(which='major',linestyle='--',linewidth='0.2',color='gray')
pl.grid(which='minor',linestyle=':',linewidth='0.1',color='gray')
pl.legend(loc='upper right')
pl.xlabel('Periodo(s)')
pl.ylabel('Acc Se(g)')
pl.show()
```
